```python
import pandas as pd
import numpy as np
import panel as pn
pn.extension('tabulator')

import hvplot.pandas
```

```python
if 'data' not in pn.state.cache.keys():

    df = pd.read_csv('Country wise GDP from 1994 to 2017.csv')

    pn.state.cache['data'] = df.copy()

else: 

    df = pn.state.cache['data']
```

```python
year_slider = pn.widgets.IntSlider(name='Year slider', start=1994, end=2012, step=3, value=1994)
year_slider
```

```python
yaxis_co2 = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['Population', 'GDP per capita (in USD)',],
    button_type='success'
)
```

```python
print(df['Country'].unique())
```

```python
# Make DataFrame Pipeline Interactive
idf = df.interactive()
```

```python
continents = ['United States', 'China', 'Japan', 'Germany', 'India', 'United Kingdom', 'Russia', 'Australia']

gdp_pipeline = (
    idf[
        (idf.Year <= year_slider) &
        (idf.Country.isin(continents))
    ]
    .groupby(['Country', 'Year'])[yaxis_co2].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='Year')  
    .reset_index(drop=True)
)
```

```python
gdp_pipeline
```

```python
gdp_plot = gdp_pipeline.hvplot(x = 'Year', by='Country', y=yaxis_co2,line_width=2, title="GDP of each country")
gdp_plot
```

```python
gdp_table = gdp_pipeline.pipe(pn.widgets.Tabulator, pagination='remote', page_size = 10, sizing_mode='stretch_width') 
gdp_table
```

```python
population_vs_gdp_scatterplot_pipeline = (
    idf[
        (idf.Year == year_slider) &
        (~ (idf.Country.isin(continents)))
    ]
    .groupby(['Country', 'Year', 'GDP change (%)'])['Pop. change (%)'].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='Year')  
    .reset_index(drop=True)
)
```

```python
population_vs_gdp_scatterplot_pipeline
```

```python
population_vs_gdp_scatterplot = population_vs_gdp_scatterplot_pipeline.hvplot(x='GDP change (%)', 
                                                                y='Pop. change (%)', 
                                                                by='Country', 
                                                                size=80, kind="scatter", 
                                                                alpha=0.7,
                                                                legend=False, 
                                                                height=500, 
                                                                width=500)
population_vs_gdp_scatterplot
```

```python
yaxis_co2_source = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['GDP change (%)', 'Pop. change (%)'], 
    button_type='success'
)

countries = ['United States', 'China', 'Japan', 'Germany', 'India', 'United Kingdom', 'Russia', 'Australia']

GDP_source_bar_pipeline = (
    idf[
        (idf.Year == year_slider) &
        (idf.Country.isin(countries))
    ]
    .groupby(['Year', 'Country'])[yaxis_co2_source].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='Year')  
    .reset_index(drop=True)
)
```

```python
GDP_source_bar_plot = GDP_source_bar_pipeline.hvplot(kind='bar', 
                                                     x='Country', 
                                                     y=yaxis_co2_source, 
                                                     title='GDP source by continent')
GDP_source_bar_plot
```

```python
#Layout using Template
template = pn.template.FastListTemplate(
    title='Country wise GDP from year 1994 to 2017', 
    sidebar=[pn.pane.Markdown("# GDP of the countries"), 
             pn.pane.Markdown("#### Gross domestic product (GDP) is the standard measure of the value added created through the production of goods and services in a country during a certain period. As such, it also measures the income earned from that production, or the total amount spent on final goods and services (less imports)."), 
             pn.pane.PNG('10698.png', sizing_mode='scale_both'),
             pn.pane.Markdown("## Settings"),   
             year_slider],
    main=[pn.Row(pn.Column(yaxis_co2, 
                           gdp_plot.panel(width=700), margin=(0,25)), 
                 gdp_table.panel(width=500)), 
          pn.Row(pn.Column(population_vs_gdp_scatterplot.panel(width=600), margin=(0,25)), 
                 pn.Column(yaxis_co2_source, GDP_source_bar_plot.panel(width=600)))],
    accent_base_color="#88d8b0",
    header_background="#88d8b0",
)
# template.show()
template.servable();
```
